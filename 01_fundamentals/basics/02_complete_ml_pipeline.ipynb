```python
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()

print("üì• LOADING DATASET")
print("="*60)

print(f"\nüìä Dataset Name: Iris Flower Classification")
print(f"Total Samples: {len(iris.data)}")
print(f"Number of Features: {len(iris.feature_names)}")
print(f"Number of Classes: {len(iris.target_names)}")

print(f"\nüå∏ Features (X):")
for i, feature in enumerate(iris.feature_names, 1):
    print(f"  {i}. {feature}")

print(f"\nüéØ Target Classes (y):")
for i, target in enumerate(iris.target_names, 1):
    print(f"  {i}. {target}")

# Convert to DataFrame for better understanding
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

print("\nüìã First 5 rows of data:")
print(df.head())

print("\nüí° Each row = One flower measurement")
print("üí° We want to predict 'species' based on 4 measurements")
```

```python
print("üîç SEPARATING FEATURES AND TARGET")
print("="*60)

# STEP 1: Get Features (X)
X = iris.data  # All measurements (4 columns)

print("\nüìè FEATURES (X) - Input to model:")
print(f"  Shape: {X.shape}")
print(f"  Meaning: {X.shape[0]} samples, {X.shape[1]} features")
print(f"\n  First sample (one flower):")
print(f"  {X[0]}")
print(f"  [sepal_length, sepal_width, petal_length, petal_width]")

# STEP 2: Get Target (y)
y = iris.target  # Species labels (0, 1, or 2)

print("\nüéØ TARGET (y) - What we want to predict:")
print(f"  Shape: {y.shape}")
print(f"  Meaning: {y.shape[0]} labels (one for each flower)")
print(f"\n  First 10 labels:")
print(f"  {y[:10]}")
print(f"  0 = setosa, 1 = versicolor, 2 = virginica")

print("\n" + "="*60)
print("‚úÖ X (features) and y (target) are now separated!")
print("‚úÖ X will be used to make predictions")
print("‚úÖ y is what we're trying to predict")
```

```python
from sklearn.model_selection import train_test_split

print("‚úÇÔ∏è  SPLITTING DATA INTO TRAIN AND TEST")
print("="*60)

# THE MAGIC FUNCTION
X_train, X_test, y_train, y_test = train_test_split(
    X,                    # Features
    y,                    # Target
    test_size=0.2,        # 20% for testing (0.2 = 20%)
    random_state=42,      # For reproducibility (same split every time)
    stratify=y            # Keep class distribution same in train & test
)

print("\nüìä ORIGINAL DATA:")
print(f"  Total samples: {len(X)}")
print(f"  Total labels: {len(y)}")

print("\n‚úÇÔ∏è  AFTER SPLIT:")
print("-" * 60)
print("TRAINING SET:")
print(f"  X_train shape: {X_train.shape}")
print(f"  ‚Üí {X_train.shape[0]} samples with {X_train.shape[1]} features")
print(f"  y_train shape: {y_train.shape}")
print(f"  ‚Üí {y_train.shape[0]} labels")
print(f"  Percentage: {len(X_train)/len(X)*100:.1f}%")

print("\nTESTING SET:")
print(f"  X_test shape: {X_test.shape}")
print(f"  ‚Üí {X_test.shape[0]} samples with {X_test.shape[1]} features")
print(f"  y_test shape: {y_test.shape}")
print(f"  ‚Üí {y_test.shape[0]} labels")
print(f"  Percentage: {len(X_test)/len(X)*100:.1f}%")
print("-" * 60)

print("\nüí° WHAT HAPPENED?")
print("  ‚úì sklearn randomly selected 20% data for testing")
print("  ‚úì Remaining 80% kept for training")
print("  ‚úì random_state=42 ensures same split every time you run")
print("  ‚úì stratify=y keeps class balance in both sets")
```

```python
print("‚öôÔ∏è  TRAIN_TEST_SPLIT PARAMETERS EXPLAINED")
print("="*60)

parameters = {
    "X, y": "Your features and target data",
    
    "test_size": """
    ‚Ä¢ How much data for testing
    ‚Ä¢ test_size=0.2 ‚Üí 20% test, 80% train
    ‚Ä¢ test_size=0.3 ‚Üí 30% test, 70% train
    ‚Ä¢ Common values: 0.2, 0.25, 0.3
    """,
    
    "random_state": """
    ‚Ä¢ A seed number for reproducibility
    ‚Ä¢ Same number = same split every time
    ‚Ä¢ Different number = different split
    ‚Ä¢ Any integer works (42 is popular!)
    ‚Ä¢ Without it, split changes each run
    """,
    
    "stratify": """
    ‚Ä¢ Keeps class distribution balanced
    ‚Ä¢ stratify=y ‚Üí same % of each class in train & test
    ‚Ä¢ Important for imbalanced datasets
    ‚Ä¢ Optional but recommended
    """
}

for param, explanation in parameters.items():
    print(f"\n{param}:")
    print(explanation.strip())
    print("-" * 60)
```

```python
print("üîç VERIFYING CLASS DISTRIBUTION")
print("="*60)

# Check original distribution
original_dist = pd.Series(y).value_counts().sort_index()

# Check train distribution
train_dist = pd.Series(y_train).value_counts().sort_index()

# Check test distribution
test_dist = pd.Series(y_test).value_counts().sort_index()

# Create comparison table
comparison = pd.DataFrame({
    'Class': iris.target_names,
    'Original': original_dist.values,
    'Train': train_dist.values,
    'Test': test_dist.values
})

comparison['Train %'] = (comparison['Train'] / comparison['Train'].sum() * 100).round(1)
comparison['Test %'] = (comparison['Test'] / comparison['Test'].sum() * 100).round(1)

print("\nüìä CLASS DISTRIBUTION COMPARISON:")
print(comparison.to_string(index=False))

print("\n‚úÖ Notice: Train and Test have similar percentages!")
print("   This is because we used stratify=y")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

print("ü§ñ TRAINING THE MODEL")
print("="*60)

# STEP 1: Create model
model = RandomForestClassifier(n_estimators=100, random_state=42)
print("\n‚úÖ Step 1: Model created (Random Forest)")

# STEP 2: Train on TRAINING data only
print("\nüéì Step 2: Training model...")
print(f"  Using X_train: {X_train.shape}")
print(f"  Using y_train: {y_train.shape}")

model.fit(X_train, y_train)

print("  ‚úÖ Training complete!")

print("\nüí° WHAT HAPPENED?")
print("  ‚Üí Model learned patterns from 120 flower samples")
print("  ‚Üí Model learned: 'If petal_length > X and petal_width < Y, then...'")
print("  ‚Üí Model has NEVER seen the test data yet!")
```

```python
print("üîÆ MAKING PREDICTIONS ON TEST DATA")
print("="*60)

# Predict on TEST data (unseen data)
y_pred = model.predict(X_test)

print(f"\n‚úÖ Predictions made on {len(X_test)} test samples")

print("\nüìä FIRST 10 PREDICTIONS vs ACTUAL:")
print("-" * 60)
comparison_df = pd.DataFrame({
    'Sample': range(1, 11),
    'Predicted': [iris.target_names[p] for p in y_pred[:10]],
    'Actual': [iris.target_names[a] for a in y_test[:10]],
    'Correct?': ['‚úÖ' if p == a else '‚ùå' for p, a in zip(y_pred[:10], y_test[:10])]
})
print(comparison_df.to_string(index=False))

print("\nüí° Model made predictions on data it NEVER saw during training!")
```

```python
print("üìä MODEL EVALUATION")
print("="*60)

# Calculate accuracy on TEST data
test_accuracy = accuracy_score(y_test, y_pred)

print(f"\nüéØ Test Accuracy: {test_accuracy * 100:.2f}%")
print(f"   Correct predictions: {int(test_accuracy * len(y_test))} out of {len(y_test)}")
print(f"   Wrong predictions: {len(y_test) - int(test_accuracy * len(y_test))}")

# Also check training accuracy (just for comparison)
y_train_pred = model.predict(X_train)
train_accuracy = accuracy_score(y_train, y_train_pred)

print(f"\nüìö Train Accuracy: {train_accuracy * 100:.2f}%")

print("\n" + "="*60)
print("üí° INTERPRETATION:")
if test_accuracy > 0.95:
    print("  ‚úÖ Excellent! Model generalizes well to new data")
elif test_accuracy > 0.90:
    print("  ‚úÖ Good! Model performs well on unseen data")
else:
    print("  ‚ö†Ô∏è  Model might need improvement")

if abs(train_accuracy - test_accuracy) < 0.05:
    print("  ‚úÖ Train and test accuracy are similar (good sign!)")
else:
    print("  ‚ö†Ô∏è  Large gap between train and test (possible overfitting)")
```

```python
import matplotlib.patches as mpatches

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('üéØ Complete ML Pipeline - Train-Test Split', 
             fontsize=18, fontweight='bold', y=0.98)

# Plot 1: Data Split Visualization
ax1 = axes[0, 0]
train_bar = ax1.barh(['Data'], [len(X_train)], color='#4ECDC4', label='Training')
test_bar = ax1.barh(['Data'], [len(X_test)], left=[len(X_train)], 
                    color='#FF6B6B', label='Testing')
ax1.set_xlabel('Number of Samples')
ax1.set_title('1Ô∏è‚É£ Data Split (80-20)', fontweight='bold')
ax1.legend()
ax1.set_xlim(0, len(X))

# Add text
ax1.text(len(X_train)/2, 0, f'{len(X_train)}\n(80%)', 
         ha='center', va='center', fontweight='bold', fontsize=12)
ax1.text(len(X_train) + len(X_test)/2, 0, f'{len(X_test)}\n(20%)', 
         ha='center', va='center', fontweight='bold', fontsize=12)

# Plot 2: Training Process
ax2 = axes[0, 1]
ax2.text(0.5, 0.7, 'üéì TRAINING PHASE', ha='center', fontsize=16, fontweight='bold')
ax2.text(0.5, 0.5, f'Model learns from\n{len(X_train)} samples', 
         ha='center', fontsize=13, bbox=dict(boxstyle='round', facecolor='#4ECDC4', alpha=0.5))
ax2.text(0.5, 0.3, f'Train Accuracy:\n{train_accuracy*100:.2f}%', 
         ha='center', fontsize=14, fontweight='bold')
ax2.axis('off')
ax2.set_title('2Ô∏è‚É£ Model Training', fontweight='bold')

# Plot 3: Testing Process
ax3 = axes[1, 0]
ax3.text(0.5, 0.7, 'üß™ TESTING PHASE', ha='center', fontsize=16, fontweight='bold')
ax3.text(0.5, 0.5, f'Model evaluated on\n{len(X_test)} NEW samples', 
         ha='center', fontsize=13, bbox=dict(boxstyle='round', facecolor='#FF6B6B', alpha=0.5))
ax3.text(0.5, 0.3, f'Test Accuracy:\n{test_accuracy*100:.2f}%', 
         ha='center', fontsize=14, fontweight='bold')
ax3.axis('off')
ax3.set_title('3Ô∏è‚É£ Model Evaluation', fontweight='bold')

# Plot 4: Results Comparison
ax4 = axes[1, 1]
accuracies = [train_accuracy * 100, test_accuracy * 100]
bars = ax4.bar(['Training', 'Testing'], accuracies, 
               color=['#4ECDC4', '#FF6B6B'], edgecolor='black', linewidth=2)
ax4.set_ylabel('Accuracy (%)', fontweight='bold')
ax4.set_ylim(0, 105)
ax4.set_title('4Ô∏è‚É£ Performance Comparison', fontweight='bold')
ax4.grid(axis='y', alpha=0.3)

# Add value labels
for i, (bar, acc) in enumerate(zip(bars, accuracies)):
    ax4.text(bar.get_x() + bar.get_width()/2, acc + 2, 
             f'{acc:.1f}%', ha='center', fontweight='bold', fontsize=12)

plt.tight_layout()
plt.show()

print("‚úÖ Complete pipeline visualized!")
```

```python
print("üìö COMPLETE ML PIPELINE - SUMMARY")
print("="*70)

steps = [
    {
        'step': '1Ô∏è‚É£ LOAD DATA',
        'code': 'X, y = iris.data, iris.target',
        'explanation': 'Get features (X) and target (y)'
    },
    {
        'step': '2Ô∏è‚É£ SPLIT DATA',
        'code': 'X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)',
        'explanation': '80% training, 20% testing'
    },
    {
        'step': '3Ô∏è‚É£ CREATE MODEL',
        'code': 'model = RandomForestClassifier()',
        'explanation': 'Choose your algorithm'
    },
    {
        'step': '4Ô∏è‚É£ TRAIN MODEL',
        'code': 'model.fit(X_train, y_train)',
        'explanation': 'Model learns from training data'
    },
    {
        'step': '5Ô∏è‚É£ MAKE PREDICTIONS',
        'code': 'y_pred = model.predict(X_test)',
        'explanation': 'Test on unseen data'
    },
    {
        'step': '6Ô∏è‚É£ EVALUATE',
        'code': 'accuracy = accuracy_score(y_test, y_pred)',
        'explanation': 'Check how well model performed'
    }
]

for item in steps:
    print(f"\n{item['step']}")
    print(f"  Code: {item['code']}")
    print(f"  ‚Üí {item['explanation']}")
    print("-" * 70)

print("\nüéØ WHY THIS PROCESS?")
print("  ‚úì Training on one set, testing on another = honest evaluation")
print("  ‚úì Prevents overfitting (memorization)")
print("  ‚úì Tells us if model will work in real world")

print("\n‚ö†Ô∏è  CRITICAL RULES:")
print("  1. NEVER use test data during training")
print("  2. NEVER peek at test data before final evaluation")
print("  3. Split data BEFORE any preprocessing")
print("  4. Use stratify for balanced classes")
print("  5. Set random_state for reproducibility")
```

```python
print("üìä COMMON TRAIN-TEST SPLIT RATIOS")
print("="*70)

split_ratios = [
    {'ratio': '80-20', 'test_size': 0.20, 'use_case': 'Most common, balanced'},
    {'ratio': '75-25', 'test_size': 0.25, 'use_case': 'More test data for better evaluation'},
    {'ratio': '70-30', 'test_size': 0.30, 'use_case': 'When you have lots of data'},
    {'ratio': '90-10', 'test_size': 0.10, 'use_case': 'When data is limited'},
]

print(f"\n{'Ratio':<12} {'test_size':<12} {'Use Case'}")
print("-" * 70)
for item in split_ratios:
    print(f"{item['ratio']:<12} {item['test_size']:<12} {item['use_case']}")

print("\nüí° CHOOSING THE RIGHT RATIO:")
print("  ‚Ä¢ More training data = Model learns better")
print("  ‚Ä¢ More test data = Better evaluation reliability")
print("  ‚Ä¢ Balance depends on total dataset size")
print("  ‚Ä¢ For small datasets (<1000): use 80-20 or 70-30")
print("  ‚Ä¢ For large datasets (>10000): can use 90-10 or 95-5")
```

```python
print("üèãÔ∏è PRACTICE: TRY DIFFERENT SPLITS")
print("="*70)

test_sizes = [0.1, 0.2, 0.3, 0.4]
results = []

for test_size in test_sizes:
    # Split data
    X_tr, X_te, y_tr, y_te = train_test_split(
        X, y, 
        test_size=test_size, 
        random_state=42, 
        stratify=y
    )
    
    # Train model
    mdl = RandomForestClassifier(random_state=42)
    mdl.fit(X_tr, y_tr)
    
    # Evaluate
    y_p = mdl.predict(X_te)
    acc = accuracy_score(y_te, y_p)
    
    results.append({
        'Train-Test Ratio': f"{int((1-test_size)*100)}-{int(test_size*100)}",
        'Train Samples': len(X_tr),
        'Test Samples': len(X_te),
        'Test Accuracy': f"{acc*100:.2f}%"
    })

results_df = pd.DataFrame(results)
print("\nüìä RESULTS WITH DIFFERENT SPLITS:")
print(results_df.to_string(index=False))

# Visualize
plt.figure(figsize=(10, 6))
plt.plot([r['Train Samples'] for r in results], 
         [float(r['Test Accuracy'].strip('%')) for r in results],
         marker='o', linewidth=2, markersize=10, color='#4ECDC4')
plt.xlabel('Training Samples', fontweight='bold')
plt.ylabel('Test Accuracy (%)', fontweight='bold')
plt.title('Effect of Train-Test Split on Accuracy', fontweight='bold', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\nüí° OBSERVATION:")
print("  ‚Üí More training data generally helps")
print("  ‚Üí But diminishing returns after certain point")
```
